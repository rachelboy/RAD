```python
import pandas
import numpy as np
```

```python
#read in our data as a pandas DataFrame (essentially a table)
data = pandas.read_csv("table1_data_raw.csv")
```

```python
#check how many non-null vlaues we have for each year for each indicator
#group by series code (the indicator type)
#then use aggregate to count the number of non-null values (not x!=x is the best non-null check I've found)
data.groupby("Series Code").aggregate(lambda col: len([x for x in col if not x!=x]))
```

```python
#select only rows with data about the poverty headcount ratio at $2/day
poverty = data[data["Series Code"]=="SI.POV.2DAY"]
#remove null values (since % will always be >=0, we use this as the non-null check)
#keep only the columns for 2005 and the country code
poverty = poverty[poverty["2005 [YR2005]"]>=0][["Country Code","2005 [YR2005]"]]
print poverty
```

```python
#select only rows with a CPIA indicator (all of these start with IQ.CPA)
#group by country code
#take the mean of all CPIA indicators as the CPIA indicator
CPIA = data[data["Series Code"].apply(lambda x: str(x)[:6]=="IQ.CPA")].groupby("Country Code").aggregate(np.mean)
#remove null values (since % will always be >=0, we use this as the non-null check)
#since we did a groupby, CPIA is indexed by the country codes, so we only need to keep the row with the data
CPIA = CPIA[CPIA["2005 [YR2005]"]>0]["2005 [YR2005]"]
print CPIA
```

```python
#select only rows with data about the poverty headcount ratio at $2/day
GDP = data[data["Series Code"]=="NY.GDP.PCAP.KN"]
#take the average yearly per captia GDP percent growth rate between 2005 and 2010
years = ["2005 [YR2005]","2006 [YR2006]","2007 [YR2007]","2008 [YR2008]","2009 [YR2009]","2010 [YR2010]"]
GDP["annual GDP growth"] = GDP.apply(lambda row: np.mean([(row[years[i+1]]-row[years[i]])/row[years[i]] for i in range(len(years)-1)]),axis=1)
#keep only the columns for 2005 and the country code
GDP = GDP[["Country Code","annual GDP growth"]]
print GDP
```

```python
#join poverty data with CPIA data
table = poverty.join(CPIA, on="Country Code", how="left", lsuffix="poverty", rsuffix="CPIA")
print table
```

```python
#join poverty and CPIA data with GDP growth data
table = table.merge(GDP, how="inner",on="Country Code")
print table
```

```python
#calculate the project poverty in 2012 with a elasticity of poverty of 2
table["poverty 2010"] = table.apply(lambda row: row["2005 [YR2005]poverty"]*((1-(2*row["annual GDP growth"]))**5),axis=1)
print table
```

```python
#get the actual poverty rate in 2010 to compare
pov_2010 = data[data["Series Code"]=="SI.POV.2DAY"][["Country Code","2010 [YR2010]"]]
table.merge(pov_2010,on="Country Code",how="inner")
```
