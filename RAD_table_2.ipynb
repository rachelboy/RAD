```python
import pandas
import numpy as np
import matplotlib
%matplotlib inline
```

```python
data = pandas.read_csv("table2_data_raw.csv")
#Look at all the data we have!
print "Series Code \t\t Series Name"
for series_name, series_code in data.groupby(["Series Name","Series Code"]).groups.keys():
    print str(series_code) + "\t\t" + str(series_name)
```

```python
#weird, what's that nan doing in our data?  Let's take a look
print data[data["Series Code"]!=data["Series Code"]]
```

```python
#Those look silly. Let's remove them.
print len(data.index)
data = data[[not x!=x for x in data["Series Code"]]]
print len(data.index)
#OK, now, how many non-null values do we have in each column?
def non_nulls(df,agg_col):
    return df.groupby(agg_col).aggregate(lambda col: len([x for x in col if not x!=x]))
non_nulls(data, "Series Code")
```

```python
#It looks like most of our CPIA data doesn't exist before 2005 - lets just drop those years
#(   we'll leave 2004 in case we want to use it as a baseline)
#It also looks like the net official aid recieved (3rd and 4th rows above) won't be 
#    very useful in that range of years, so we'll drop that too
drop_years = [str(y) for y in range(1984,2004)]
drop_cols = [y+" [YR"+y+"]" for y in drop_years]
less_data = data.drop(drop_cols, axis=1) # axis=1 means columns

drop_rows_1 = less_data[less_data["Series Code"]=="DT.ODA.OATL.CD"].index
drop_rows_2 = less_data[less_data["Series Code"]=="DT.ODA.OATL.KD"].index
drop_rows = drop_rows_1+drop_rows_2
less_data.drop(drop_rows,inplace=True) #axis=0 means drop rows, 
                                #inplace=True means it drops them in place instead of returning a new dataframe
#OK, cool, lets look at our non-nulls again
non_nulls(less_data, "Series Code")
```

```python
#It looks like most of our CPIA data doesn't exist before 2005 - lets just use 2005-2012

def get_col_years(years):
    return [y+" [YR"+y+"]" for y in [str(x) for x in years]]

def all_valid(vals):
    out = False
    for val in vals:
        out = out or val!=val
    return not out

#Lovely! Let's start computing things
#The paper uses four-year aggregates, so we will too. one from 2005-2008, one from 2009-2012
rows_list = []
grouped = less_data.groupby(["Country Code"])
for country, cols in grouped.groups.items():
    row = {"Country Code":country}
    temp = less_data.loc[cols]
    
    #get the CPIA
    CPIAs = np.mean(temp[temp["Series Code"].apply(lambda x: str(x)[:6]=="IQ.CPA")])
    CPIA_05_08 = np.mean([CPIAs[y] for y in get_col_years(range(2005,2009))])
    CPIA_09_12 = np.mean([CPIAs[y] for y in get_col_years(range(2009,2013))])
    
    temp.set_index("Series Code",inplace=True)
    #get GDP and growth
    GDP_2004 = temp.loc["NY.GDP.PCAP.KN"]["2004 [YR2004]"]
    GDP_2008 = temp.loc["NY.GDP.PCAP.KN"]["2008 [YR2008]"]
    growth_05_08 = (GDP_2008/GDP_2004) - 1
    growth_09_12 = (temp.loc["NY.GDP.PCAP.KN"]["2012 [YR2012]"]/GDP_2008)-1
    
    #get aid
    aid = temp.loc["DT.ODA.ALLD.KD"]
    GDP = temp.loc["NY.GDP.PCAP.KN"]
    coef_aid_05_08 = np.mean([aid[y]/GDP[y] for y in get_col_years(range(2005,2009))])
    coef_aid_09_12 = np.mean([aid[y]/GDP[y] for y in get_col_years(range(2009,2013))])
    
    #construct the row
    bool1 = all_valid([CPIA_05_08, GDP_2004, growth_05_08, coef_aid_05_08])
    bool2 = all_valid([CPIA_09_12, GDP_2008, growth_09_12, coef_aid_09_12])
    if bool1 or bool2:
        row["GDP per cap 2004"] = GDP_2004 if bool1 else 0
        row["GDP growth 2005-2008"] = growth_05_08 if bool1 else 0
        row["CPIA 2005-2008"] = CPIA_05_08 if bool1 else 0
        row["coef of aid 2005-2008"] = coef_aid_05_08 if bool1 else 0


        row["GDP per cap 2008"] = GDP_2008 if bool2 else 0
        row["GDP growth 2009-2012"] = growth_09_12 if bool2 else 0
        row["CPIA 2009-2012"] = CPIA_09_12 if bool2 else 0
        row["coef of aid 2009-2012"] = coef_aid_09_12 if bool2 else 0

        rows_list.append(row)
df = pandas.DataFrame(rows_list)
df.set_index("Country Code",inplace=True)

#Annnnnd we have a dataframe!  How nice.  Let's look at it!
df
```
