```python
import pandas
import numpy as np
import scipy as sp
import matplotlib.pyplot as pp
import thinkstats2 as ts
import thinkplot as tp
%matplotlib inline
```

```python
data = pandas.read_csv("table2_data_raw.csv")
#Look at all the data we have!
print "Series Code \t\t Series Name"
for series_name, series_code in data.groupby(["Series Name","Series Code"]).groups.keys():
    print str(series_code) + "\t\t" + str(series_name)
```

```python
#weird, what's that nan doing in our data?  Let's take a look
print data[data["Series Code"]!=data["Series Code"]]
```

```python
#Those look silly. Let's remove them.
print len(data.index)
data = data[[not x!=x for x in data["Series Code"]]]
print len(data.index)
#OK, now, how many non-null values do we have in each column?
def non_nulls(df,agg_col):
    return df.groupby(agg_col).aggregate(lambda col: len([x for x in col if not x!=x]))
non_nulls(data, "Series Code")
```

```python
#It looks like most of our CPIA data doesn't exist before 2005 - lets just drop those years
#(   we'll leave 2004 in case we want to use it as a baseline)
#It also looks like the net official aid recieved (3rd and 4th rows above) won't be 
#    very useful in that range of years, so we'll drop that too
drop_years = [str(y) for y in range(1984,2004)]
drop_cols = [y+" [YR"+y+"]" for y in drop_years]
less_data = data.drop(drop_cols, axis=1) # axis=1 means columns

drop_rows_1 = less_data[less_data["Series Code"]=="DT.ODA.OATL.CD"].index
drop_rows_2 = less_data[less_data["Series Code"]=="DT.ODA.OATL.KD"].index
drop_rows = drop_rows_1+drop_rows_2
less_data.drop(drop_rows,inplace=True) #axis=0 means drop rows, 
                                #inplace=True means it drops them in place instead of returning a new dataframe
#OK, cool, lets look at our non-nulls again
non_nulls(less_data, "Series Code")
```

```python
#It looks like most of our CPIA data doesn't exist before 2005 - lets just use 2005-2012

def get_col_years(years):
    return [y+" [YR"+y+"]" for y in [str(x) for x in years]]

def all_valid(vals):
    out = False
    for val in vals:
        out = out or val!=val
    return not out

#Lovely! Let's start computing things
#The paper uses four-year aggregates, so we will too. one from 2005-2008, one from 2009-2012
rows_list = []
grouped = less_data.groupby(["Country Code"])
for country, cols in grouped.groups.items():
    
    temp = less_data.loc[cols]
    
    #get the CPIA
    CPIAs = np.mean(temp[temp["Series Code"].apply(lambda x: str(x)[:6]=="IQ.CPA")])
    CPIA_05_08 = np.mean([CPIAs[y] for y in get_col_years(range(2005,2009))])
    CPIA_09_12 = np.mean([CPIAs[y] for y in get_col_years(range(2009,2013))])
    
    temp.set_index("Series Code",inplace=True)
    #get GDP and growth
    GDP_2004 = temp.loc["NY.GDP.PCAP.KN"]["2004 [YR2004]"]
    GDP_2008 = temp.loc["NY.GDP.PCAP.KN"]["2008 [YR2008]"]
    growth_05_08 = (GDP_2008/GDP_2004) - 1
    growth_09_12 = (temp.loc["NY.GDP.PCAP.KN"]["2012 [YR2012]"]/GDP_2008)-1
    
    #get aid
    aid = temp.loc["DT.ODA.ALLD.KD"]
    GDP = temp.loc["NY.GDP.PCAP.KN"]
    #you may notice that aid is in constant 2011 USD, while GDP is in constant LCU
    #this means our coefficient of aid will be off by some constant (what ever the conversion from 2011 USD to LCU is)
    #don't worry, we'll take care of this when we normalize things!
    aid_05_08 = np.sum([aid[y] for y in get_col_years(range(2005,2009))])
    GDP_05_08 = np.sum([GDP[y] for y in get_col_years(range(2005,2009))])
    coef_aid_05_08 = aid_05_08/GDP_05_08

    aid_09_12 = np.sum([aid[y] for y in get_col_years(range(2009,2013))])
    GDP_09_12 = np.sum([GDP[y] for y in get_col_years(range(2009,2013))])
    coef_aid_09_12 = aid_09_12/GDP_09_12
    
    #construct the row
    bool1 = all_valid([CPIA_05_08, GDP_2004, growth_05_08, coef_aid_05_08])
    bool2 = all_valid([CPIA_09_12, GDP_2008, growth_09_12, coef_aid_09_12])
    if bool1:
        row = {"Country Code":country}
        row["initial GDP"] = GDP_2004
        row["GDP growth"] = growth_05_08
        row["CPIA"] = CPIA_05_08
        row["coef of aid"] = coef_aid_05_08
        rows_list.append(row)
        
    if bool2:
        row = {"Country Code":country}
        row["initial GDP"] = GDP_2008
        row["GDP growth"] = growth_09_12
        row["CPIA"] = CPIA_09_12
        row["coef of aid"] = coef_aid_09_12
        rows_list.append(row)
df = pandas.DataFrame(rows_list)
df.set_index("Country Code",inplace=True)

#Annnnnd we have a dataframe!  How nice.  Let's look at it!
df
```

```python
#Wow! So many numbers! What do we DO with them all?
#Let's look at them first . We're looking for relationships to growth, so let's start there.

def residuals(guess,actual):
    res = 0
    for y1,y2 in zip(guess,actual):
        res += (y1-y2)**2
    return res

#we'll plot the values against each other
xs, ys = df["CPIA"].values, df["GDP growth"].values
pp.scatter(xs, ys)
pp.ylabel("GDP growth")
pp.xlabel("CPIA")
#And we'll get the correlation coefficient
print "Coefficient of correlation:", np.corrcoef(xs, ys)[0,1]

#And let's look at how well a line could fit it
mean = np.mean(ys)
pp.plot([1.5,4.5],[mean,mean],color = "red",label = "mean")

coefs = np.lib.polyfit(xs,ys,1)
fit_ys = np.lib.polyval(coefs, xs)
print "Line of best fit:", coefs[0], "+", coefs[1], "*x"
pp.plot(xs,fit_ys, color="green",label = "least squares")
pp.legend()

resd_guess = residuals([mean]*len(ys),ys)
resd_fit = residuals(fit_ys,ys)

print "Residuals around mean:",resd_guess
print "Residuals around line of best fit:", resd_fit
```

```python
#In fact, we can do the same thin with all fo our variables:

def investigate_fit(xlabel,ylabel, data):
    #we'll plot the values against each other
    xs, ys = data[xlabel].values, df[ylabel].values
    pp.scatter(xs, ys)
    pp.ylabel(ylabel)
    pp.xlabel(xlabel)
    #And we'll get the correlation coefficient
    print "Coefficient of correlation:", np.corrcoef(xs, ys)[0,1]

    #And let's look at how well a line could fit it
    mean = np.mean(ys)
    pp.plot([min(xs),max(xs)],[mean,mean],color = "red",label = "mean")

    coefs = np.lib.polyfit(xs,ys,1)
    fit_ys = np.lib.polyval(coefs, xs)
    print "Line of best fit:", coefs[0], "+", coefs[1], "*x"
    pp.plot(xs,fit_ys, color="green",label = "least squares")
    pp.legend()

    resd_guess = residuals([mean]*len(ys),ys)
    resd_fit = residuals(fit_ys,ys)

    print "Residuals around mean:",resd_guess
    print "Residuals around line of best fit:", resd_fit
    
print "Coefficent of Aid\n"
investigate_fit("coef of aid","GDP growth",df)
```

```python
#And for initial GDP
print "Initial GDP\n"
investigate_fit("initial GDP","GDP growth",df)
```

```python
#Allllllright. It's time for linear regression. Let's see how the individual distributions look.

def plot_hist(dataframe, col_label, bin_factor = 10.0, **kwargs):
    #first we bin the values
    bin_factor = float(bin_factor)
    vals_list = [int(x*bin_factor)/bin_factor for x in dataframe[col_label].values]
    #then we make the histogram
    hist = ts.Hist()
    for val in vals_list:
        hist.Incr(val)
    #then we plot it!
    tp.Hist(hist,**kwargs)
    pp.legend()
    
plot_hist(df,"CPIA",label="CPIA")
# you can see below, that's not the worst distribution to work with. 
# We might as well scale it from 0-1, but that's it.
df["CPIA scaled"] = df["CPIA"].apply(lambda x: x/6.0) #the actual range for CPIA is from 1 to 6
plot_hist(df, "CPIA scaled", bin_factor=30, label="CPIA scaled", color="red")
```

```python
plot_hist(df,"GDP growth", bin_factor=20)
#This is also reasonable looking, and it's already roughly scaled from negative one to one.
#We'll leave it alone
df["GDP growth scaled"] = df["GDP growth"]
```

```python
#Just looking at the numbers above, we know coef of aid is waaaay to large to work with
#Let's scale it before we do anything
max_val = np.max(df["coef of aid"])
df["coef of aid scaled"] = df["coef of aid"].apply(lambda x: x/max_val)
plot_hist(df,"coef of aid scaled", bin_factor=30, label = "aid scaled 0-1")
```

```python
#That distribution is pretty intensely bunhced up on one side. That might not have been the right scale.
# Let's try putting the median at .02, instead
med_val = np.median(df["coef of aid"])
df["coef of aid scaled"] = df["coef of aid"].apply(lambda x: x/(50*med_val))
plot_hist(df,"coef of aid scaled", bin_factor=20, label = "aid scaled 0.1", color="red")
#Our real goal is to get the constants on about the same scale, so that's probably good enought to be getting on with
```

```python
#That's also really skewed, so let's spread it out a little
df["initial GDP scaled"] = df["initial GDP"].apply(lambda x: 10*x/max_val)
plot_hist(df,"initial GDP scaled", bin_factor=10, label = "GDP linearly scaled")

#This one may also be improved with a log distribution (shown in red) (though again, I don't think the paper uses it)
plot_hist(df,"initial GDP log scaled", bin_factor=10, label = "GDP log scaled", color="red")
```
